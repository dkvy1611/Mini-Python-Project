```python
# Import các library cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tải lên dataset
df = pd.read_csv(r'C:\Users\ADMIN\Downloads\DA_Excel.csv')

# Xem qua một số dòng đầu của dataset
df.head()
```

```python
# Xóa các cột không dùng
df = df.drop(columns=['Ngành hàng cấp 1','Ngành hàng cấp 2','Ngành hàng cấp 3','Số lượt xem sản phẩm','Số đánh giá'])
df.head()
```

```python
# Check data type
df.dtypes
```

```python
# Đổi data type cột Giá sang dạng float để tiện cho tính toán theo yêu cầu
df['Giá'] = df['Giá'].astype(float)
df.dtypes
```

```python
## Lập bảng thống kê số đã bán, doanh số theo từng hãng điện thoại?

# Định nghĩa hàm thay đổi giá trị của cột Thương hiệu
def update_brand(row):
    if pd.isna(row['Thương hiệu']) or row['Thương hiệu'] == '':
        if 'Xiaomi' in row['Tên sản phẩm']:
            return 'xiaomi'
        elif 'SS' in row['Tên sản phẩm'] or 'Samsung' in row['Tên sản phẩm']:
            return 'samsung'
        elif 'nokia' in row['Tên sản phẩm']:
            return 'nokia'
        else:
            return 'Khác'
    else:
        return row['Thương hiệu']

# Áp dụng hàm update_brand vào DataFrame
df['Thương hiệu'] = df.apply(update_brand, axis=1)

# Thiết lập tùy chọn hiển thị số của pandas để không hiện chữ e
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Tính toán cột Doanh số bằng cách nhân Giá với Số đã bán
df['Doanh số'] = df['Giá'] * df['Số đã bán']

# Tính tổng doanh số và số lượng bán theo Thương hiệu
result1 = df[df['Thương hiệu'] != ''].groupby('Thương hiệu').agg({'Doanh số': 'sum', 'Số đã bán': 'sum'}).reset_index()
print(result1.sort_values(by = 'Doanh số', ascending = False))
```

```python
# Hàm để xác định nước của từng thương hiệu
def classify_country(brand):
    if brand in ['apple', 'google', 'xsmart']:
        return 'Mỹ'
    elif brand in ['xiaomi', 'vivo', 'oppo', 'realme', 'poco', 'redmi', 'tecno', 'infinix', 'xiaomi youpin', 'kudixiong', 'itel']:
        return 'Trung Quốc'
    elif brand in ['samsung', 'lg']:
        return 'Hàn Quốc'
    else:
        return 'Khác'

# Thêm cột "Nước" dựa trên thương hiệu
df['Nước'] = df['Thương hiệu'].apply(classify_country)

# Tính cột "Doanh_số" bằng cách nhân cột "Giá" với cột "Số đã bán"
df['Doanh số'] = df['Giá'] * df['Số đã bán']

# Tính tổng doanh số và tổng số đã bán theo nước
result2 = df.groupby('Nước').agg({'Doanh số': 'sum', 'Số đã bán': 'sum'}).reset_index()
result2_sdb = result2.sort_values(by='Số đã bán', ascending=False)

# Điện thoại của nước nào bán chạy nhất? (số đã bán cao nhất) - Trung Quốc
print(result2_sdb)
```

```python
# Điện thoại của nước nào có doanh số cao nhất? - Mỹ
result2_ds = result2.sort_values(by='Doanh số', ascending=False)
print(result2_ds)
```
